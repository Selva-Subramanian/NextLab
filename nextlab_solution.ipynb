```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/NextLab/Datasets/browser_rankings_data.csv")
```

```python
df = df.dropna(axis=0)
df = df.reset_index()
df = df.drop(['index'], axis = 1)
```

```python
def remove_punctuation(text):
    '''a function for removing punctuation'''
    import string
    translator = str.maketrans('', '', string.punctuation)
    # return the text stripped of punctuation marks
    return text.translate(translator)
```

```python
df["Long Description"] = df["Long Description"].apply(remove_punctuation)
df.head(10)
```

```python
df['Keyword Presence'] = np.nan
for i in range(len(df)):
  keyword = df['Keyword'][i].split(' ')
  for j in keyword:
    if j.title() not in keyword:
      keyword.append(j.title())
      result = []
  for k in keyword:
    if k in df['Long Description'][i].split(' ')[:10]:
      result.append(True)
    else:
      result.append(False)
  df['Keyword Presence'][i] = any(result)
```

```python
df[df['Keyword Presence'] == True]['Rank'].value_counts()[:5].plot(kind='bar')
plt.show()
```

```python
df[df['Keyword Presence'] == False]['Rank'].value_counts()[:5].plot(kind='bar')
plt.show()
```

#### The presence of keyword in the description has no correaltion with ranking.

```python
from happytransformer import HappyTextToText
from happytransformer import TTSettings
```

```python
happy_text = HappyTextToText("T5", "vennify/t5-base-grammar-correction")
```

```python
args = TTSettings(num_beams=5, min_length=1)
```

```python
result = happy_text.generate_text("grammar: This sentences has has bads grammar.", args=args)
```

```python
result.text
```

```python
def corrected(text):
  txt = "grammar: "+str(text)
  result = happy_text.generate_text(txt, args=args)
  return result.text
```

```python
df['corrected'] = np.nan
for i in range(len(df)):
  if len(df['text'][i].split(' '))>3:
    text = df['text'][i]
    df['corrected'][i] = corrected(text)
  else:
    text = df['text'][i]
    df['corrected'][i] = df['text'][i]
```

```python
def similarity(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union
    
df['grammatical accuracy'] = np.nan
for i in range(len(df)):
  df['grammatical accuracy'][i] = round(similarity(df['text'][i].split(' '),df['corrected'][i].split(' ')),2)
```
